## Reading the meta data data file

First download the data from [this google folder](https://drive.google.com/drive/folders/1FgAS3UU0EfdKVqxgSc3gn6nHh84aC89-?usp=share_link). This includes a spreadsheet and pickle file (of the same data).

```python
import pandas as pd

df = pd.read_pickle('./data/Kenya_hdx_tabular_data_metadata.pkl')
```

```python
print(df.columns)
display(df)
```

As a reminder, [HDX](https://data.humdata.org/) has datasets (basically the pages you see there), and then each file is a 'resource'. Fields from these are included in the table, but there are some fields which weren't parsed out. You can access the full dataset and resource objects records for each using these columns. As they are not true JSON, you will need to use `ast.literal_eval` to convert them todictionaries.

```python
import ast

ast.literal_eval(df.loc[10, 'dataset'])

ast.literal_eval(df.loc[10, 'resource'])
```

The data contains excerpts of the first few rows. The data in the table is a string, so it needs to be converted to JSON then pandas. 

**Note:** This is the unprocessed table data. For Excel files it can be a bit strange due to blank header rows and merged cells

```python
import json

display(pd.DataFrame(json.loads(df.loc[0, 'first_nrows'])))
```
